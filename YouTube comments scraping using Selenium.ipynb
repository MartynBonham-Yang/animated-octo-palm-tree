```python
# set options to be headless, ..
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
#Import other selenium packages necessary 
import time
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
import pandas as pd


pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
```

```python


#Clean up & remove new lines (represented by \n) and replace with spaces
def remove_nl(comment):
  comment = comment.str.replace('\n', ' ')
  return comment

#Define webdriver operation function:
def CommentScraper(YouTubeLink, howMany):

  data = []
  wd = webdriver.Chrome('chromedriver',options=options)
  with wd as driver: 
    wait = WebDriverWait(driver,15)
    driver.get(YouTubeLink)

    for item in range(howMany): 
        wait.until(EC.visibility_of_element_located((By.TAG_NAME, "body"))).send_keys(Keys.END)
        time.sleep(12)

    for comment in wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "#content"))):
        data.append(comment.text)

  df = pd.DataFrame(data, columns = ['comment'])
  df.drop(df.index[0], inplace = True)
  df_no_blanks = df[df['comment']!='']
  df_no_blanks = df_no_blanks.reset_index().drop('index', axis = 1)
  comments = df_no_blanks.drop(index=0)
  comments['comment'] = remove_nl(comments['comment'])

  return comments
```

```python
df = CommentScraper('https://www.youtube.com/watch?v=Zz5HLec0vaQ', 10)
```
